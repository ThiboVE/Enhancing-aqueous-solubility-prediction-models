```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_full = pd.read_csv("metrics_full.csv")
df_rdkit = pd.read_csv("metrics_rdkit.csv")
```

```python
compressed_df_full = df_full.groupby("epoch").apply(lambda x: x.bfill().iloc[0])[:-1]
compressed_df_full.head()
```

```python
compressed_df_rdkit = df_rdkit.groupby("epoch").apply(lambda x: x.bfill().iloc[0])[:-1]
compressed_df_rdkit.head()
```

```python
plt.plot(compressed_df_full.index, compressed_df_full["train_loss_epoch"], color="tab:blue", linestyle="--")
plt.plot(compressed_df_full.index, compressed_df_full["val_loss"], color="tab:blue", label="full")

plt.plot(compressed_df_rdkit.index, compressed_df_rdkit["train_loss_epoch"], color="tab:red", linestyle="--")
plt.plot(compressed_df_rdkit.index, compressed_df_rdkit["val_loss"], color="tab:red", label="rdkit")

plt.ylim(0, 0.3)

plt.legend()
```
